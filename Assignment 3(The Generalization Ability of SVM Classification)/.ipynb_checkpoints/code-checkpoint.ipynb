```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.utils import shuffle

df=pd.read_csv('/content/sample_data/letter-recognition.csv')
diag_map = {'A': 1.0, 'B': -1.0,'C': 1.0, 'D': -1.0,'E': 1.0, 'F': -1.0,'G': 1.0, 'H': -1.0,'I': 1.0, 'J': -1.0,'K': 1.0, 'L': -1.0,'M': 1.0, 'N': -1.0
            ,'O': 1.0, 'P': -1.0,'Q': 1.0, 'R': -1.0,'S': 1.0, 'T': -1.0,'U': 1.0, 'V': -1.0,'W': 1.0, 'X': -1.0,
            'Y': 1.0, 'Z': -1.0}
df['diagnosis'] = df['diagnosis'].map(diag_map)
df

X=df.drop('diagnosis',axis=1)
y=df['diagnosis']
m=8000
X=X[:m]
y=y[:m]
print(X.shape)

df1=df
df1=df1.sample(frac=1)
X1=df1.drop('diagnosis',axis=1)[:m]
y1=df1['diagnosis'][:m]

tr_si=int(0.8*m)
X_train=X1[:tr_si]
y_train=y1[:tr_si]
X_test=X1[tr_si:]
y_test=y1[tr_si:]

print(X_train.shape,X_test.shape)

from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)

y_pred=clf.predict(X_test)

"""**RESULTS SVM WITHOUT MARKOV SAMPLING**"""

print("accuracy on test dataset: {}".format(accuracy_score(y_test, y_pred)))
print("recall on test dataset: {}".format(recall_score(y_test, y_pred)))
print("precision on test dataset: {}".format(recall_score(y_test, y_pred)))

N1=1000
X_val=X[:N1]
y_val=y[:N1]
print(y_val.shape)

from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X_val, y_val)

y_pre=clf.predict(X_test)
print(y_pre)

def Cost(Y_pred, Y):
    # calculate hinge loss
    if Y_pred*Y>1.0:
        return 0
    return 1.0-Y_pred*Y

X=X.to_numpy()
y=y.to_numpy()

Ni=0
j=0
k=2
q=1/1.2
print(q)
m=y.shape[0]
print(m)

import math
mp=0
mn=0
ct=0
X_training=[]
Y_training=[]
X_testing=[]
Y_testing=[]
X_t=X[0].reshape(1,-1)
print(X_t)
y_t=y[0].reshape(1,-1)[0]
X_training.append(X_t[0])
Y_training.append(y_t[0])
if m%2==0 and y_t[0]==1.0:
    mp+=1
elif m%2==0 and y_t[0]==-1.0:
    mn+=1
pred_t=clf.predict(X_t)
print(pred_t)
p=1999999.0
for i in range(1,m):
    X_z=X[i].reshape(1,-1)
    y_z=y[i].reshape(1,-1)[0]
    pred_z=clf.predict(X_z)
    P=math.exp(-Cost(pred_z[0],y_z[0]))/math.exp(-Cost(pred_t[0],y_t[0]))
    if ct>=k:
        Pdd=q*P
        if Pdd>0.5:
            X_training.append(X_z[0])
            Y_training.append(y_z[0])
            ct+=1 
            X_t=X_z
            y_t=y_z
            pred_t=pred_z
            Ni+=1
        else:
            X_testing.append(X_z[0])
            Y_testing.append(y_z[0])
            ct=0
    elif P==1.0 and y_z==y_t:
        Pd=math.exp(-y_z[0]*pred_z[0])/math.exp(-y_t[0]*pred_t[0])
        if Pd>0.5:
            X_training.append(X_z[0])
            Y_training.append(y_z[0])
            ct+=1
            X_t=X_z
            y_t=y_z
            pred_t=pred_z
            Ni+=1
        else:
            X_testing.append(X_z[0])
            Y_testing.append(y_z[0])
            ct=0
    elif P==1.0 and y_z[0]*y_t[0]==-1.0 or P<1.0:
        Pd=math.exp(-y_z[0]*pred_z[0])/math.exp(-y_t[0]*pred_t[0])
        if Pd>0.5:
            X_training.append(X_z[0])
            Y_training.append(y_z[0])
            ct+=1
            X_t=X_z
            y_t=y_z
            pred_t=pred_z
            Ni+=1
        else:
            X_testing.append(X_z[0])
            Y_testing.append(y_z[0])
            ct=0
    elif P>1.0:
        X_training.append(X_z[0])
        Y_training.append(y_z[0])
        ct+=1
        X_t=X_z
        y_t=y_z
        pred_t=pred_z
        Ni+=1
    if Ni>=N1:
        clf.fit(np.array(X_training),np.array(Y_training))
        j+=1
    if j>=k:
        break

X_training=np.array(X_training)
Y_training=np.array(Y_training)
X_testing=np.array(X_testing)
Y_testing=np.array(Y_testing)

print(X_training.shape[0],Y_testing.shape[0])

clf = svm.SVC(kernel='linear')
clf.fit(X_training, Y_training)

y_pred=clf.predict(X_test)
print("accuracy on test dataset: {}".format(accuracy_score(y_test, y_pred)))
print("recall on test dataset: {}".format(recall_score(y_test, y_pred)))
print("precision on test dataset: {}".format(recall_score(y_test, y_pred)))
```
